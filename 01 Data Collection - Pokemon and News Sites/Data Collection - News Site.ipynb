### Import libraries

```python
from dotenv import load_dotenv
load_dotenv()
import os

import requests
import json 

from bs4 import BeautifulSoup
```

### Get URL and API key for NewsAPI

```python
url = 'http://newsapi.org/v2/everything?'
```

### Check available sources

```python
sources = 'http://newsapi.org/v2/sources?apiKey=c2fd8f8c95834e54ac52747f0df8318a'
requests.get(sources).json()
```

### Set parameters for NewsAPI

```python
parameters_0311_0312 = {
    'domains' : 'mtv.com',
    'apiKey' : os.getenv("key"),
    'from' : '2021-03-11',
    'to' : '2021-03-12',
    'language' : 'en'
}

parameters_0311_0312_json = requests.get(url, params = parameters_0311_0312)
parameters_0311_0312_json = parameters_0311_0312_json.json()

parameters_0311_0312_json
```

### Check keys of news JSON 

```python
parameters_0311_0312_json.keys()
extract_articles = parameters_0311_0312_json['articles']
```

### Store essential data to `results`

```python
results = []

for i in range(len(extract_articles)):
    # Extract full article
    article = requests.get(extract_articles[i]['url'])
    soup = BeautifulSoup(article.content, 'html.parser')
    article_content = soup.find_all("section", {"class": "entry-content"})
    article_content_text = article_content[0].get_text().replace('\n', '')
    
    # Store temporarily in a dictionary
    results_dict = {}
    results_dict['date'] = extract_articles[i]['publishedAt']
    results_dict['title'] = extract_articles[i]['title']
    results_dict['full_article'] = article_content_text
    results_dict['author'] = extract_articles[i]['author']
    
    # Append to list
    results.append(results_dict)

results
```

### Save as JSON file

```python
with open("NewsSite_MTV.json", "w") as outfile: 
    json.dump(results, outfile) 
```
